```python
import pandas as pd
import matplotlib.pyplot as plt
mushrooms = pd.read_csv('mushrooms.csv')
mushrooms.head()
mushrooms.dtypes
```

```python
# convertig the mushroom dataframe datypes to category from object dtype
cols = mushrooms.select_dtypes(["object"]).columns
mushrooms[cols] = mushrooms[cols].astype('category')
print(mushrooms[cols])
mushrooms.dtypes
```

```python
# grouping the categorical mushroom dataframe by class column and assign it to edibleclass
edibleclass=mushrooms.groupby(['class']).count()
edibleclass
```

```python
# plottig a pie chart of population column of the edibleclass with labels of 'Edible' and 'Poisonous'
labels=['Edible','Poisonous']
plt.pie(edibleclass['population'],labels=labels,autopct='%.1f %%')
plt.title('Edible?')
plt.show()
```

```python
# grouping the categorical mushroom dataframe by habitat column and assign it to haitat
habitat=mushrooms.groupby(['habitat']).count()
habitat
```

```python
# plottig a donut chart of gruopby habitat column with labels of 'Grasses','Leaves','Meadows','Paths','Urban','Waste' and 'Wood'
labels=['Grasses','Leaves','Meadows','Paths','Urban','Waste','Wood']

plt.pie(habitat['class'], labels=labels,
        autopct='%1.1f%%', pctdistance=0.85)
  
center_circle = plt.Circle((0, 0), 0.40, fc='white')
fig = plt.gcf()

fig.gca().add_artist(center_circle)
  
plt.title('Mushroom Habitats')
  
plt.show()
```

```python
capcolor=mushrooms.groupby(['cap-color']).count()
capcolor
```
